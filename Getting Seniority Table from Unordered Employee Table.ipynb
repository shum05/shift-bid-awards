```python
import pandas as pd

# Load the data from employee_table.csv
df = pd.read_csv('data/employee_table.csv')

df.head()
```

```python
df.dtypes
```

```python
# Ensure hire_date is in datetime format
df['hire_date'] = pd.to_datetime(df['hire_date'])
df.head(10)
```

```python
df.dtypes
```

```python

# Sort by hire_date to determine the seniority order and reset the index
df1 = df.sort_values(by='hire_date').reset_index(drop=True)
df1.head(10)
```

```python
# Add seniority_number column based on sorted order
df1['seniority_number'] = range(1, len(df) + 1)
df1 = df1[['employee_number', 'seniority_number', 'hire_date']]
df1.head(10)
```

```python

# Save the updated DataFrame to seniority_table.csv
df1.to_csv('data/seniority_table.csv', index=False)

# Display the updated DataFrame
print(df1)
```
